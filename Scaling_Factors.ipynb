IMPORTING ORIGINAL FOLDERS

```python

import os
import pandas as pd

# Define the root directory where your folders are located
root_directory = 'C:/Users/Rachel/OneDrive - UCLA IT Services/Documents/DWP_Electrification_Cal/data/data/'

# Create an empty dictionary to store the DataFrames
df_dict = {}

# Iterate over each folder
folders = os.listdir(root_directory)
for folder_name in folders:
    folder_path = os.path.join(root_directory, folder_name)
    
    # Check if the path is a directory
    if os.path.isdir(folder_path):
    
        # Get a list of files in the folder
        files = os.listdir(folder_path)
        
        # Iterate over each file in the folder
        for file_name in files:
            file_path = os.path.join(folder_path, file_name)
            
            # Check if the path is a file
            if os.path.isfile(file_path):
            
                # Read the file as a DataFrame and store it in the dictionary
                key = os.path.join(folder_name, file_name)
                df = pd.read_csv(file_path)
        
                
                # Select specific columns
                selected_columns = ['upgrade', 'in.geometry_building_type_recs', 'timestamp',
                                   'units_represented','out.site_energy.total.energy_consumption.kwh',
                                   'out.electricity.total.energy_consumption.kwh','out.fuel_oil.total.energy_consumption.kwh',
                                   'out.natural_gas.total.energy_consumption.kwh','out.propane.total.energy_consumption.kwh',
                                    'out.load.cooling.energy_delivered.kbtu','out.load.heating.energy_delivered.kbtu',
                                   'out.load.hot_water.energy_delivered.kbtu','out.emissions.all_fuels.lrmer_95decarbby2035_15_2025start.co2e_kg',
                                   'out.emissions.all_fuels.lrmer_lowrecost_15_2025start.co2e_kg','out.emissions.all_fuels.lrmer_lowrecost_25_2025start.co2e_kg',
                                   'out.emissions.all_fuels.lrmer_midcase_15_2025start.co2e_kg']
                df = df[selected_columns]
               
                # Accounting for units represented to get load of single unit
                df['elec.consumed'] = df['out.electricity.total.energy_consumption.kwh'] / df['units_represented']
                df['nat_gas.consumed'] = df['out.natural_gas.total.energy_consumption.kwh'] / df['units_represented']
                df['total.consumed'] = df['out.site_energy.total.energy_consumption.kwh'] / df['units_represented']
                df['elec.load'] = df['elec.consumed'] / .25
                df['nat_gas.load'] = df['nat_gas.consumed'] / .25
                df['total_load'] = df['total.consumed'] / .25
                df['elec_sum'] = sum(df['elec.consumed']) 
                df['nat_gas_sum'] = sum(df['nat_gas.consumed']) 
                df['total_sum'] = sum(df['total.consumed']) 
                df_dict[key] = df
```

```python
import numpy as np
keys = list(df_dict.keys())
sums = [[''] * 4 for _ in range(55)]
```

```python
count = 0
for x in keys:
    df = df_dict[x]
    sums[count][0] = x
    sums[count][1] = df['elec_sum'][0]
    sums[count][2] = df['nat_gas_sum'][0]
    sums[count][3] = df['total_sum'][0]
    count = count+1
```

```python
from itertools import product

names = [row[0] for row in sums]
elec_sums = [row[1] for row in sums]
nat_gas_sums = [row[2] for row in sums]
total_sums = [row[3] for row in sums]
names = list(product(names, repeat=2))
elec_combos = list(product(elec_sums, repeat=2))
nat_gas_combos = list(product(nat_gas_sums, repeat=2))
total_combos = list(product(total_sums, repeat=2))
```

```python
both_names = [row[0] + '\\' + row[1] for row in names]
elec_multipliers = [row[1] / row[0] for row in elec_combos]
nat_gas_multipliers = [row[1] / row[0] for row in nat_gas_combos]
total_multipliers = [row[1] / row[0] for row in total_combos]
```

```python
##all_multipliers = np.array([both_names, elec_multipliers, nat_gas_multipliers, total_multipliers]).T
```

```python
split_columns = [row.split('\\') for row in both_names]
column1, column2,column3,column4 = zip(*split_columns)
combined_array = np.array([column1, column2, column3, column4, elec_multipliers,nat_gas_multipliers, total_multipliers]).T
headers = ['starting_upgrade','starting_home','other_upgrade','other_home','elec_mult','nat_gas_mult','total_mult']
mult_df = pd.DataFrame(combined_array, columns=headers)
mult_df['starting_home_cut'] = mult_df['starting_home'].str[8:]
mult_df['other_home'] = mult_df['other_home'].str[8:]
filtered_df = mult_df[mult_df['starting_home_cut'] == mult_df['other_home']]
```

```python
# Concatenate columns 1 and 2
concat = filtered_df['starting_upgrade'] +"\\" + filtered_df['starting_home'] + "\\" + filtered_df['other_upgrade']
new_array = np.column_stack((concat, filtered_df['elec_mult'], filtered_df['nat_gas_mult'], filtered_df['total_mult']))
headers = ['name','elec_mult','nat_gas_mult','total_mult']
mult_df = pd.DataFrame(new_array, columns=headers)
```
